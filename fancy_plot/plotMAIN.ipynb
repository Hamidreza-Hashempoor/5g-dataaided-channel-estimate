```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python

def PlotMSE(SNRlist,
             LSmselist,
               LMMSEmselist,
                 ChannelNetmselist,
                   ConvLMMSEmselist,
                     HYBRIDMLP0mselist,
                       HYBRIDMLP1mselist,
                         HYBRIDSVM0mselist,
                           HYBRIDSVM1mselist,
                             CENetmselist,
                               LSLSmselist,
                                 Langevianmselist,
                                   JCdmselist,
                                     DDCEmselist):
    matplotlib.colors.ColorConverter.colors['z'] = (0.99,0.4,0.0)
    matplotlib.colors.ColorConverter.colors['q'] = (0.39,0.1,0.0)
    matplotlib.colors.ColorConverter.colors['w'] = (0.5,0.9,0.9)
    matplotlib.colors.ColorConverter.colors['u'] = (0.22,0.7,0.9)
    matplotlib.colors.ColorConverter.colors['f'] = (0.15,0.4,0.6)
    matplotlib.colors.ColorConverter.colors['a'] = (0.85,0.55,0.3)
    matplotlib.colors.ColorConverter.colors['e'] = (0.10,0.95,0.5)
    matplotlib.colors.ColorConverter.colors['i'] = (0.15,0.35,0.95)
    matplotlib.colors.ColorConverter.colors['j'] = (0.65,0.15,0.25)
    plt.figure(figsize=(5,7))
    plt.plot(SNRlist, LSmselist, "--", label="LS ", marker='o', linewidth=1)
    plt.plot(SNRlist, LSmselist, "ob")
    plt.plot(SNRlist, LMMSEmselist, "-r", label="LMMSE ", marker='s', linewidth=1)
    plt.plot(SNRlist, LMMSEmselist, "sr")
    plt.plot(SNRlist, ChannelNetmselist, "-w", label="ChannelNet ", marker='^', linewidth=1)
    plt.plot(SNRlist, ChannelNetmselist, "^w")
    plt.plot(SNRlist, ConvLMMSEmselist, "-g", label="MAP Data-Aided ", marker='d', linewidth=1)
    plt.plot(SNRlist, ConvLMMSEmselist, "dg")
    
    plt.plot(SNRlist, CENetmselist, "--a", label="CENet ", marker='<', linewidth=1)
    plt.plot(SNRlist, CENetmselist, "<a")
    plt.plot(SNRlist, Langevianmselist, "--e", label="Langevian ", marker='<', linewidth=1)
    plt.plot(SNRlist, Langevianmselist, "<e")
    plt.plot(SNRlist, JCdmselist, "--i", label="JCD ", marker='<', linewidth=1)
    plt.plot(SNRlist, JCdmselist, "<i")
    plt.plot(SNRlist, DDCEmselist, "--j", label="DDCE ", marker='<', linewidth=1)
    plt.plot(SNRlist, DDCEmselist, "<j")

    
    
    plt.plot(SNRlist, HYBRIDMLP0mselist, "-z", label="hybrid 0/MLP ", marker='o', linewidth=1)
    plt.plot(SNRlist, HYBRIDMLP0mselist, "oz")
    plt.plot(SNRlist, HYBRIDMLP1mselist, "--u", label="hybrid 1/MLP", marker='v', linewidth=1)
    plt.plot(SNRlist, HYBRIDMLP1mselist, "vu")
    plt.plot(SNRlist, HYBRIDSVM0mselist, "--q", label="hybrid 0/SVM ", marker='x', linewidth=1)
    plt.plot(SNRlist, HYBRIDSVM0mselist, "xq")
    plt.plot(SNRlist, HYBRIDSVM1mselist, "--f", label="hybrid 1/SVM ", marker='<', linewidth=1)
    plt.plot(SNRlist, HYBRIDSVM1mselist, "<f")
    
    

    plt.grid(True, which='both', axis='y', ls="-")
    plt.grid(True, which='major', axis='x', ls="-")
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.yscale('log')
    plt.xlabel("SNR", fontsize=15)
    plt.ylabel("MSE", fontsize=15)
    plt.legend(loc="best",  fontsize=10)
    plt.savefig('SNRMSE.pdf', bbox_inches='tight')
    plt.show()
    
def PlotBER(SNRlist, LSmselist,  LMMSEberlist, ChannelNetbellist, ConvLMMSEberlist, HYBRIDMLP0berlist, HYBRIDMLP1berlist, HYBRIDSVMb0erlist, HYBRIDSVMb1erlist, CNetberlist, LSLSberlist, Langevianberlist, JCdberlist, DDCEberlist):
    matplotlib.colors.ColorConverter.colors['z'] = (0.99,0.4,0.0)
    matplotlib.colors.ColorConverter.colors['q'] = (0.39,0.1,0.0)
    matplotlib.colors.ColorConverter.colors['w'] = (0.5,0.9,0.9)
    matplotlib.colors.ColorConverter.colors['u'] = (0.22,0.7,0.9)
    matplotlib.colors.ColorConverter.colors['f'] = (0.15,0.4,0.6)
    matplotlib.colors.ColorConverter.colors['a'] = (0.85,0.55,0.3)
    matplotlib.colors.ColorConverter.colors['e'] = (0.10,0.95,0.5)
    matplotlib.colors.ColorConverter.colors['i'] = (0.9,0.2,0.95)
    matplotlib.colors.ColorConverter.colors['j'] = (0.3,0.6,0.35)
    matplotlib.colors.ColorConverter.colors['k'] = (0.45,0.95,0.50)
    plt.figure(figsize=(5,7))
    plt.plot(SNRlist, LSLSberlist, "--e", label="LS + ML ", marker='<', linewidth=1)
    plt.plot(SNRlist, LSLSberlist, "<e")
    plt.plot(SNRlist, LSmselist, "--", label="LS", marker='o', linewidth=1)
    plt.plot(SNRlist, LSmselist, "ob")
    plt.plot(SNRlist, LMMSEberlist, "-r", label="LMMSE ", marker='s', linewidth=1)
    plt.plot(SNRlist, LMMSEberlist, "sr")
    plt.plot(SNRlist, ChannelNetbellist, "-w", label="ChannelNet ", marker='^', linewidth=1)
    plt.plot(SNRlist, ChannelNetbellist, "^w")
    
    plt.plot(SNRlist, ConvLMMSEberlist, "-g", label="MAP Data-Aided ", marker='d', linewidth=1)
    plt.plot(SNRlist, ConvLMMSEberlist, "dg")
    
    plt.plot(SNRlist, CNetberlist, "--a", label="CENet ", marker='<', linewidth=1)
    plt.plot(SNRlist, CNetberlist, "<a")
    
    plt.plot(SNRlist, Langevianberlist, "--i", label="Langevian ", marker='<', linewidth=1)
    plt.plot(SNRlist, Langevianberlist, "<i")
    plt.plot(SNRlist, JCdberlist, "--j", label="JCD ", marker='<', linewidth=1)
    plt.plot(SNRlist, JCdberlist, "<j")
    plt.plot(SNRlist, DDCEberlist, "--k", label="DDCE ", marker='<', linewidth=1)
    plt.plot(SNRlist, DDCEberlist, "<k")
    
    plt.plot(SNRlist, HYBRIDMLP0berlist, "-z", label="hybrid 0/MLP ", marker='o', linewidth=1)
    plt.plot(SNRlist, HYBRIDMLP0berlist, "oz")
    plt.plot(SNRlist, HYBRIDMLP1berlist, "--u", label="hybrid 1/MLP", marker='v', linewidth=1)
    plt.plot(SNRlist, HYBRIDMLP1berlist, "vu")
    plt.plot(SNRlist, HYBRIDSVMb0erlist, "--q", label="hybrid 0/SVM ", marker='x', linewidth=1)
    plt.plot(SNRlist, HYBRIDSVMb0erlist, "xq")
    plt.plot(SNRlist, HYBRIDSVMb1erlist, "--f", label="hybrid 1/SVM ", marker='<', linewidth=1)
    plt.plot(SNRlist, HYBRIDSVMb1erlist, "<f")
    
    
    
    
    plt.grid(True, which='both', axis='y', ls="-")
    plt.grid(True, which='major', axis='x', ls="-")
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.yscale('log')
    plt.xlabel("SNR", fontsize=15)
    plt.ylabel("BER", fontsize=15)
    plt.legend(loc="best",  fontsize=10)
    plt.savefig('SNRBER.pdf', bbox_inches='tight')
    plt.show()
    
snr_list = [ -10, -5, 0, 5,10,15,20]

LSmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
LSLSmselist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))

ChannelNetmselist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
CENetmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))

LMMSEmselist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
ConvLMMSEmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
Langevianmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
JCdmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
DDCEmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))

HYBRIDMLP0mselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
HYBRIDMLP1mselist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))


HYBRIDSVM0mselist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
HYBRIDSVM1mselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))

PlotMSE(snr_list, LSmselist, LMMSEmselist, ChannelNetmselist, ConvLMMSEmselist, HYBRIDMLP0mselist, HYBRIDMLP1mselist, HYBRIDSVM0mselist, HYBRIDSVM1mselist, CENetmselist, LSLSmselist, Langevianmselist, JCdmselist, DDCEmselist)

LSberlist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
LSLSberlist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
ChannelNetmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
CNetberlist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))


LMMSEberlist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
ConvLMMSEberlist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
Langevianberlist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
JCdberlist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
DDCEberlist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))


HYBRIDMLP0berlist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
HYBRIDMLP1berlist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))



HYBRIDSVM0berlist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))
HYBRIDSVM1berlist = list( np.random.choice(np.linspace(0, 0.1, 10000), size=7, replace=False))

PlotBER(snr_list, LSberlist, LMMSEberlist, ChannelNetmselist, ConvLMMSEberlist, HYBRIDMLP0berlist, HYBRIDMLP1berlist, HYBRIDSVM0berlist, HYBRIDSVM1berlist, CNetberlist, LSLSberlist, Langevianberlist, JCdberlist, DDCEberlist)
```
