```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python



def PlotMSE(SNRlist,
            LSmselist,
            LMMSEmselist,
            ChannelNetmselist,
            ConvLMMSEmselist,
            HYBRIDMLP0mselist,
            HYBRIDMLP1mselist,
            HYBRIDSVM0mselist,
            HYBRIDSVM1mselist,
            CENetmselist,
            Langevianmselist,
            jcdmselist,
            DDCHEmselist):
    matplotlib.colors.ColorConverter.colors['z'] = (0.99,0.4,0.0)
    matplotlib.colors.ColorConverter.colors['q'] = (0.39,0.1,0.0)
    matplotlib.colors.ColorConverter.colors['w'] = (0.5,0.9,0.9)
    matplotlib.colors.ColorConverter.colors['u'] = (0.22,0.7,0.9)
    matplotlib.colors.ColorConverter.colors['f'] = (0.15,0.4,0.6)
    matplotlib.colors.ColorConverter.colors['a'] = (0.85,0.55,0.3)
    matplotlib.colors.ColorConverter.colors['e'] = (0.10,0.95,0.5)
    matplotlib.colors.ColorConverter.colors['i'] = (0.15,0.35,0.95)
    matplotlib.colors.ColorConverter.colors['j'] = (0.65,0.15,0.25)
    plt.figure(figsize=(5,7))
    plt.plot(SNRlist, LSmselist, "--", label="LS ", marker='o', linewidth=1)
    plt.plot(SNRlist, LSmselist, "ob")
    plt.plot(SNRlist, LMMSEmselist, "-r", label="LMMSE ", marker='s', linewidth=1)
    plt.plot(SNRlist, LMMSEmselist, "sr")
    plt.plot(SNRlist, ChannelNetmselist, "-w", label="ChannelNet ", marker='^', linewidth=1)
    plt.plot(SNRlist, ChannelNetmselist, "^w")
    plt.plot(SNRlist, ConvLMMSEmselist, "-g", label="MAP Data-Aided ", marker='d', linewidth=1)
    plt.plot(SNRlist, ConvLMMSEmselist, "dg")
    
    
    plt.plot(SNRlist, CENetmselist, "--a", label="CENet", marker='<', linewidth=1)
    plt.plot(SNRlist, CENetmselist, "<a")
    plt.plot(SNRlist, Langevianmselist, "--e", label="Langevian", marker='<', linewidth=1)
    plt.plot(SNRlist, Langevianmselist, "<e")
    plt.plot(SNRlist, jcdmselist, "--i", label="JCD", marker='<', linewidth=1)
    plt.plot(SNRlist, jcdmselist, "<i")
    plt.plot(SNRlist, DDCHEmselist, "--j", label="DDCE", marker='<', linewidth=1)
    plt.plot(SNRlist, DDCHEmselist, "<j")
    
    plt.plot(SNRlist, HYBRIDMLP0mselist, "-z", label="hybrid 0/MLP ", marker='o', linewidth=1)
    plt.plot(SNRlist, HYBRIDMLP0mselist, "oz")
    plt.plot(SNRlist, HYBRIDMLP1mselist, "--u", label="hybrid 1/MLP", marker='v', linewidth=1)
    plt.plot(SNRlist, HYBRIDMLP1mselist, "vu")
    plt.plot(SNRlist, HYBRIDSVM0mselist, "--q", label="hybrid 0/SVM ", marker='x', linewidth=1)
    plt.plot(SNRlist, HYBRIDSVM0mselist, "xq")
    plt.plot(SNRlist, HYBRIDSVM1mselist, "--f", label="hybrid 1/SVM ", marker='<', linewidth=1)
    plt.plot(SNRlist, HYBRIDSVM1mselist, "<f")
    
    
    

    plt.grid(True, which='both', axis='y', ls="-")
    plt.grid(True, which='major', axis='x', ls="-")
    plt.ylim((5*10e-6, 10e-3))
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.yscale('log')
    plt.xlabel("RB", fontsize=15)
    plt.ylabel("MSE", fontsize=15)
    plt.legend(loc="upper right",  fontsize=10)
    plt.savefig('RBMSE.pdf', bbox_inches='tight')
    plt.show()
    
def PlotBER(SNRlist,
            LSmselist,
            LMMSEberlist,
            ChannelNetbellist,
            ConvLMMSEberlist,
            HYBRIDMLP0berlist,
            HYBRIDMLP1berlist,
            HYBRIDSVMb0erlist, 
            HYBRIDSVMb1erlist,
            CENetbellist,
            Langevianberlist,
            JCDberlist,
            LSLSberlist,
            DDCHberlist):
    matplotlib.colors.ColorConverter.colors['z'] = (0.99,0.4,0.0)
    matplotlib.colors.ColorConverter.colors['q'] = (0.39,0.1,0.0)
    matplotlib.colors.ColorConverter.colors['w'] = (0.5,0.9,0.9)
    matplotlib.colors.ColorConverter.colors['u'] = (0.22,0.7,0.9)
    matplotlib.colors.ColorConverter.colors['f'] = (0.15,0.4,0.6)
    matplotlib.colors.ColorConverter.colors['a'] = (0.85,0.55,0.3)
    matplotlib.colors.ColorConverter.colors['e'] = (0.10,0.95,0.5)
    matplotlib.colors.ColorConverter.colors['i'] = (0.15,0.35,0.95)
    matplotlib.colors.ColorConverter.colors['j'] = (0.45,0.85,0.15)
    matplotlib.colors.ColorConverter.colors['l'] = (0.65,0.15,0.25)
    plt.figure(figsize=(5,7))
    plt.plot(SNRlist, LSLSberlist, "--j", label="LS + ML ", marker='<', linewidth=1)
    plt.plot(SNRlist, LSLSberlist, "<j")
    plt.plot(SNRlist, LSmselist, "--", label="LS", marker='o', linewidth=1)
    plt.plot(SNRlist, LSmselist, "ob")
    plt.plot(SNRlist, LMMSEberlist, "-r", label="LMMSE ", marker='s', linewidth=1)
    plt.plot(SNRlist, LMMSEberlist, "sr")
    plt.plot(SNRlist, ChannelNetbellist, "-w", label="ChannelNet ", marker='^', linewidth=1)
    plt.plot(SNRlist, ChannelNetbellist, "^w")
    
    plt.plot(SNRlist, ConvLMMSEberlist, "-g", label="MAP Data-Aided ", marker='d', linewidth=1)
    plt.plot(SNRlist, ConvLMMSEberlist, "dg")
    
    plt.plot(SNRlist, CENetbellist, "--a", label="CENet ", marker='<', linewidth=1)
    plt.plot(SNRlist, CENetbellist, "<a")
    plt.plot(SNRlist, Langevianberlist, "--e", label="Langevian ", marker='<', linewidth=1)
    plt.plot(SNRlist, Langevianberlist, "<e")
    plt.plot(SNRlist, JCDberlist, "--i", label="JCD ", marker='<', linewidth=1)
    plt.plot(SNRlist, JCDberlist, "<i")
    plt.plot(SNRlist, DDCHberlist, "--l", label="DDCE ", marker='<', linewidth=1)
    plt.plot(SNRlist, DDCHberlist, "<l")
    
    
    plt.plot(SNRlist, HYBRIDMLP0berlist, "-z", label="hybrid 0/MLP ", marker='o', linewidth=1)
    plt.plot(SNRlist, HYBRIDMLP0berlist, "oz")
    plt.plot(SNRlist, HYBRIDMLP1berlist, "--u", label="hybrid 1/MLP", marker='v', linewidth=1)
    plt.plot(SNRlist, HYBRIDMLP1berlist, "vu")
    plt.plot(SNRlist, HYBRIDSVMb0erlist, "--q", label="hybrid 0/SVM ", marker='x', linewidth=1)
    plt.plot(SNRlist, HYBRIDSVMb0erlist, "xq")
    plt.plot(SNRlist, HYBRIDSVMb1erlist, "--f", label="hybrid 1/SVM ", marker='<', linewidth=1)
    plt.plot(SNRlist, HYBRIDSVMb1erlist, "<f")
    
    
    plt.grid(True, which='both', axis='y', ls="-")
    plt.grid(True, which='major', axis='x', ls="-")
    plt.ylim((3*10e-6, 10e-2))
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.yscale('log')
    plt.xlabel("RB", fontsize=15)
    plt.ylabel("BER", fontsize=15)
    plt.legend(loc="upper right",  fontsize=10)
    plt.savefig('RBBER.pdf', bbox_inches='tight')
    plt.show()
    
RB_list = [ 5,10,15,20,25,30,35,40,45,50,55, 60]
LSmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))

ChannelNetmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
CENetmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))

LMMSEmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
ConvLMMSEmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
Langevianmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
jcdmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
DDCHEmselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
HYBRIDMLP0mselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
HYBRIDMLP1mselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))

HYBRIDSVM0mselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
HYBRIDSVM1mselist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))

PlotMSE(RB_list, LSmselist, LMMSEmselist, ChannelNetmselist, ConvLMMSEmselist, HYBRIDMLP0mselist, HYBRIDMLP1mselist, HYBRIDSVM0mselist, HYBRIDSVM1mselist, CENetmselist, Langevianmselist, jcdmselist, DDCHEmselist)

LSberlist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
LSLSberlist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))

ChannelNetbellist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
CENetbellist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))

LMMSEberlist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
ConvLMMSEberlist =  list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
Langevianberlist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
JCDberlist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
DDCHberlist =  list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))

HYBRIDMLP0berlist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
HYBRIDMLP1berlist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))

HYBRIDSVMb0erlist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))
HYBRIDSVMb1erlist = list(np.random.choice(np.linspace(0, 0.1, 10000), size=12, replace=False))

PlotBER(RB_list, LSberlist, LMMSEberlist, ChannelNetbellist, ConvLMMSEberlist, HYBRIDMLP0berlist, HYBRIDMLP1berlist, HYBRIDSVMb0erlist, HYBRIDSVMb1erlist, CENetbellist, Langevianberlist, JCDberlist, LSLSberlist, DDCHberlist)
```
